```python
%%file mQ1.py
#!/usr/bin/env python

import sys
import string  

# input comes from STDIN (standard input)
i = 0
for line in sys.stdin:
    if i == 0:    
        i+=1
        pass
        
    else:
        # remove leading and trailing whitespace
        line = line.strip()
        # split the line into words
        columns = line.split(",")
        #columns[0] = state, columns[5] e 6 = coordenadas
        print(columns[24] + "\t" + "(" + columns[5] + ","+ columns[6]+")")
```

```python
%%file rQ1.py
#!/usr/bin/python

import sys
import string  

lastState = None
coordenadasSet = set()

# input comes from STDIN (standard input)
for line in sys.stdin:
    # remove leading and trailing whitespace
    line = line.strip()
    # split the line into words
    state, coordenadas = line.split("\t")
    latitude = coordenadas.split(",")[0][1:]
    longitude = coordenadas.split(",")[1][:-1]
    coordenadas = (latitude, longitude)
    
    if state == lastState:
        coordenadasSet.add(coordenadas)
    else:
        if lastState:
            print(lastState+"\t"),
            print(len(coordenadasSet))
        lastState = state
        coordenadasSet = {coordenadas}

if lastState:
    print(lastState + "\t"),
    print(len(coordenadasSet))
```

# Q2 - Which counties have the best/worst air quality? (Rank counties considering pollutants level!)


```python
%%file mQ2.py
#!/usr/bin/env python

import sys
import string  

# input comes from STDIN (standard input)
i = 0
for line in sys.stdin:
    if i == 0:    
        i+=1
        pass
        
    else:
        # remove leading and trailing whitespace
        line = line.strip()
        # split the line into words
        columns = line.split(",")
        #columns[0] = state, columns[1] = county, columns[16] = media poluicao para dia
        print(columns[0]+"|"+columns[1] + "\t" + columns[16])
```

```python
%%file rQ2.py
#!/usr/bin/python

import sys
import string  

lastStateCounty = None
polutionLevels = []

# input comes from STDIN (standard input)
for line in sys.stdin:
    # remove leading and trailing whitespace
    line = line.strip()
    # split the line into words
    stateCounty, polution = line.split("\t")
    
    if stateCounty == lastStateCounty:
        polutionLevels.append(float(polution))
    else:
        if lastStateCounty:
            print(lastStateCounty+"\t"),
            print('{0:f}'.format((sum(polutionLevels)/len(polutionLevels))))
        lastStateCounty = stateCounty
        polutionLevels = [float(polution)]

if lastStateCounty:
    print(lastStateCounty + "\t"),
    print('{0:f}'.format((sum(polutionLevels)/len(polutionLevels))))
```

## Reducer

```python
%%file reducer1_Ex3.py
#!/usr/bin/python

import sys

#fixar (ano,estado) e somar todas as qual ar
year_aux = None
state_aux = None
airqTotal = 0
total = 0


# input comes from STDIN
for line in sys.stdin:
    # remove leading and trailing whitespace
    line = line.strip()

    #print(line)
    # parse the input we got from mapper.py
    
    year = line.split('\t')[0].split(' ')[0]
    state = line.split('\t')[0].split(' ')[1]
    airq = line.split('\t')[1]
    #print(year,state,airq)
    airq = float(airq)    
   

    if (year,state) == (year_aux,state_aux):
        airqTotal += airq
        total += 1
    else:
        if year_aux:
            print(year_aux,state_aux,airqTotal/total)
            total=0
        year_aux = year
        state_aux = state
        airqTotal = airq
        total += 1
        
if year_aux:
    print(year_aux,state_aux,airqTotal/total)
```

# Q4 - For each state, what is the average distance (in km) of the monitors in that state to the state center? For simplicity, assume that 1 degree of latitude or logitude equals to 111 km. (Monitor dispersion per state!)

```python
%%file mQ4.py
#!/usr/bin/env python

import sys
import string  

# input comes from STDIN (standard input)
i=0
# Obter coordenadas do ficheiro auxiliar
for line in sys.stdin:
    if i == 0:    
        i+=1
        pass
    else:
        word=line.split(',')
        print(word[1].lower().replace(" ","")+' ('+str((float(word[2])*111+float(word[3])*111)/2)+ ','+str((float(word[4])*111+float(word[5])*111)/2)+')')
```

```python
%%file mAux2Q4.py
#!/usr/bin/env python

import sys
import string 
import pickle
import math

# input comes from STDIN (standard input)

usa_state=pickle.load( open( "usa_state.bin", "rb" ))

for line in sys.stdin:
    word=line.split()
    state=word[0]
    if(state in usa_state):
        coord_monitor=word[1][1:-1].split(',')
        latitude_monitor=float(coord_monitor[0])
        longitude_monitor=float(coord_monitor[1])
        coord_state=usa_state[state].replace('\n','')[1:-2].split(',')
        latitude_usa=float(coord_state[0])
        longitude_usa=float(coord_state[1])
        dist=math.sqrt(((latitude_monitor-latitude_usa)**2)+(longitude_monitor-longitude_usa)**2)
        print(state,' ',str(dist))
```

### Organizar estados com latitude e longitude media para o ficheiro usastates

```python
%%file mQ5.py
#!/usr/bin/env python

import sys
import string  

# input comes from STDIN (standard input)
i = 0
for line in sys.stdin:
    if i == 0:    
        i+=1
        pass
        
    else:
        # remove leading and trailing whitespace
        line = line.strip()
        # split the line into words
        columns = line.split(",")
        #words[1] => State name, #latMedia, longMedia
        print(columns[1] + "\t" + "(" + str((float(columns[2])+float(columns[3]))/2) +","+ str((float(columns[4])+float(columns[5]))/2)+")")

```

```python
%%file rQ5.py
#!/usr/bin/python

import sys
import string 
import pickle

dictStatesCoordinates = dict()
# input comes from STDIN (standard input)
for line in sys.stdin:
    # remove leading and trailing whitespace
    line = line.strip()
    
    state, coordenadas = line.split("\t")
    
    latitude = coordenadas.split(",")[0][1:].replace("'", "")
    longitude = coordenadas.split(",")[1][:-1].replace("'", "")
    coordenadas = (float(latitude), float(longitude))

    dictStatesCoordinates[state] = coordenadas
    
pickle.dump(dictStatesCoordinates, open("PickleUSAStates.bin", "wb" ) ) 
```

### Comparar as coordenadas de cada monitor com as do estado para determinar o quadrante

```python
%%file mAux2Q5.py
#!/usr/bin/env python

import sys
import string 
import pickle

dictStatesCoordinates = pickle.load( open( "PickleUSAStates.bin", "rb" ) )


# ler o ficheiro epa
for line in sys.stdin:
    # remove leading and trailing whitespace
    line = line.strip()
    # split the line into words
    
    estado = line.split("\t")[0].split("|")[0]
    coordendas = line.split("\t")[0].split("|")[1]
    
    
    tuploCoordenadas = eval(coordendas)
    coordenadas = (float(tuploCoordenadas[0]), float(tuploCoordenadas[1]))

    #percorrer todos os monitores para cada estado
    #monitor[0] = latitude
    #monitor[1] = longitude
    if estado in dictStatesCoordinates:
        if (coordenadas[0] <= float(dictStatesCoordinates[estado][0])) and (coordenadas[1] <= float(dictStatesCoordinates[estado][1])):
            print(estado+"NW"+ "\t" + str(1))

        elif (coordenadas[0] <= float(dictStatesCoordinates[estado][0])) and (coordenadas[1] >  float(dictStatesCoordinates[estado][1])):
            print(estado+"SW"+ "\t" +str(1))

        elif (coordenadas[0] > float(dictStatesCoordinates[estado][0])) and (coordenadas[1] <=  float(dictStatesCoordinates[estado][1])):
            print(estado+"NE"+ "\t"  +str(1))

        elif (coordenadas[0] > float(dictStatesCoordinates[estado][0])) and (coordenadas[1] >  float(dictStatesCoordinates[estado][1])):
            print(estado+"SE"+ "\t"  +str(1))

#fazer outro mapreduce para contar o nr de monitores 
```
